```python
nltk.download('stopwords')
```

```python
def process_text(subject):
    nopunc=[char for char in subject if char not in string.punctuation]
    nopunc=''.join(nopunc)
    
    clean_words=[word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    
    return clean_words
```

```python
df['subject'].head().apply(process_text)
```

```python
message4='hello world hello hello world play'
message5='test test test test one hello'
print(message4)
print()

from sklearn.feature_extraction.text import CountVectorizer
bow4=CountVectorizer(analyzer=process_text).fit_transform([[message4],[message5]])
print(bow4)
print()

print(bow4.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
message_bow=CountVectorizer(analyzer=process_text).fit_transform(df['subject'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(message_bow,df['subject'],test_size=0.20,random_state=0)
```

```python
message_bow.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB().fit(x_train,y_train)
```

```python
print(classifier.predict(x_train))
print(y_train.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
pred=classifier.predict(x_train)
print(classification_report(y_train,pred))
print()
print('confusion Matrix: \n',confusion_matrix(y_train,pred))
print()
print('Accuracy: ',accuracy_score(y_train,pred))
```

```python
print(classifier.predict(x_test))
print(y_test.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
pred=classifier.predict(x_test)
print(classification_report(y_test,pred))
print()
print('confusion Matrix: \n',confusion_matrix(y_test,pred))
print()
print('Accuracy: ',accuracy_score(y_test,pred))
```
